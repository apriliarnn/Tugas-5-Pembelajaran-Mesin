```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate Q1, Q3, and IQR
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Pilih hanya kolom numerik untuk perhitungan IQR
df_numerik = df.select_dtypes(include=['number'])
Q1 = df_numerik.quantile(0.25)
Q3 = df_numerik.quantile(0.75)
IQR = Q3 - Q1

# Pastikan df, Q1, dan Q3 selaras
df, Q1 = df.align(Q1, axis=1, copy=False)
df, Q3 = df.align(Q3, axis=1, copy=False)
IQR = Q3 - Q1

# Terapkan filter IQR untuk menghilangkan outlier
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Periksa bentuk data hasil
df.shape
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
# Pilih hanya kolom numerik dari df untuk menghitung korelasi
df_numerik = df.select_dtypes(include=['number'])

# Buat heatmap korelasi
plt.figure(figsize=(10,5))
c = df_numerik.corr()  # Hitung matriks korelasi hanya pada kolom numerik
sns.heatmap(c, cmap="BrBG", annot=True)
plt.show()

# Tampilkan matriks korelasi
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

```python
columns_name = df.columns

fig, ax = plt.subplots(9, 1, figsize=(10, 20))
fig.subplots_adjust(hspace=0.75)

for i in range(9):
    sns.boxplot(x=df[columns_name[i]], ax=ax[i])
    ax[i].set_title(columns_name[i])

plt.show()
```
